```python
df_train['source']='train'
df_test['source']='test'
```

```python
df = pd.concat([df_train, df_test],ignore_index=True)
df
```

```python
df.apply(lambda x: len(x.unique()))
```

```python
# Filling the null values of column Age with mean

from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy='mean')
df['Item_Weight'] = imp.fit_transform(df['Item_Weight'].values.reshape(-1,1))
impm = SimpleImputer(strategy='most_frequent')
df['Outlet_Size'] = impm.fit_transform(df['Outlet_Size'].values.reshape(-1, 1))

df
```

```python
df.isnull().sum()
```

```python
df.drop(['Item_Type','Outlet_Establishment_Year'],axis=1,inplace=True)
```

```python
for i in df.columns[:-1]:
    if df.dtypes[i]=='object':
        print('%s \n %s \n' % (i, df[i].unique()))
```

```python
df['Item_Identifier'] = df['Item_Identifier'].apply(lambda x: x[:2])
```

```python
df['Item_Fat_Content'].unique()
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF':'Low Fat', 'reg':'Regular', 'low fat':'Low Fat'})
df['Item_Fat_Content'].unique()
```

```python
# Label Encoder - converting categorical to numeric

from sklearn.preprocessing import LabelEncoder

for i in ['Item_Identifier', 'Item_Fat_Content', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']:

    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
train = df.loc[df['source']=="train"]
test = df.loc[df['source']=="test"]
```

```python
test.columns
```

```python
test.drop(['Item_Outlet_Sales','source'],axis=1,inplace=True)
train.drop(['source'],axis=1,inplace=True)
```

```python
train.to_csv("Data/train_bigdata_modified.csv", index=False)
test.to_csv("Data/test_bigdata_modified.csv", index=False)
```

```python
train_df = pd.read_csv('Data/train_bigdata_modified.csv')
test_df = pd.read_csv('Data/train_bigdata_modified.csv')
```

```python
x_train = train_df.drop(['Item_Outlet_Sales', 'Item_Identifier', 'Outlet_Identifier'], axis=1)
y_train = train_df['Item_Outlet_Sales']
```

```python
x_test = test_df.drop(['Item_Outlet_Sales', 'Item_Identifier', 'Outlet_Identifier'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print(y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(LinearRegression(), x_train, y_train, cv=20, scoring='neg_mean_squared_error')
cv_score = np.sqrt(np.abs(cv_score))
print(cv_score)
```

```python
print("CV Score : \nMean - %s \nStd - %s\nMin - %s\nMax - %s" % (np.mean(cv_score), np.std(cv_score), np.min(cv_score), np.max(cv_score)))
```
