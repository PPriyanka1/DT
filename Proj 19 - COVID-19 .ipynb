# Proj 19 - COVID19

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid = pd.read_csv('Data/aggregate_covid_country.csv', parse_dates=['Date'])
covid
```

```python
covid.isnull().sum()
```

```python
covid.dtypes
```

```python
covid.describe()
```

```python
plt.figure(figsize=(20,30))

plt.barh(covid["Country"],covid["Confirmed"],  color = 'blue')

plt.title("Countries with number of confirmed covid-19 cases ",fontsize = 20)

plt.xlabel("Confirmed cases", fontsize=10)

plt.ylabel("Countries", fontsize=10)

plt.xticks(fontsize = 8)

plt.yticks(fontsize = 10)

plt.show()
```

```python
plt.figure(figsize=(20,30))

plt.barh(covid["Country"],covid["Recovered"],  color = 'green')

plt.title("Countries with number of recovered covid-19 cases ",fontsize = 20)

plt.xlabel("Recovered cases", fontsize=10)

plt.ylabel("Countries", fontsize=10)

plt.xticks(fontsize = 8)

plt.yticks(fontsize = 10)

plt.show()
```

```python
plt.figure(figsize=(20,30))

plt.barh(covid["Country"],covid["Deaths"],  color = 'red')

plt.title("Countries with number of death covid-19 cases ",fontsize = 20)

plt.xlabel("Death cases", fontsize=10)

plt.ylabel("Countries", fontsize=10)

plt.xticks(fontsize = 8)

plt.yticks(fontsize = 10)

plt.show()
```
